# Cleaning Data

```python
#drop unrelated or blank columns
#axis-1 is defined to select whole column
df.drop(['Status','unnamed1'], axis = 1, inplace = True)
```

```python
df.info()
```

```python
#find the null values, if any
pd.isnull(df).sum()
```

```python
df.shape
```

```python
#dropping null values now
df.dropna(inplace = True)
```

```python
df.shape
```

```python
pd.isnull(df).sum()
```

# Transforming Data

```python
df['Amount'] = df['Amount'].astype('int')
#data type will be changed
```

```python
df.columns
```

```python
#rename column
df.rename(columns = {'Martial_Status':'Shaadi'})
```

```python
#to return description(count,mean etc)
df[['Age','Orders','Amount']].describe()
```

# Data Analysis

**Gender**

```python
df.columns
```

```python
ax=sns.countplot(x = 'Gender', data = df)
for bars in ax.containers:   #to display value on bar
  ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
spendthrift = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Gender', y = 'Amount', data = spendthrift)
```

*Conclusion:: Purchasing Power of women is greater than than men*



**Age**

```python
df.columns
```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
#grouped by age, for buyers
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age)
```

*Conclusion:: The most of the buyers are of group between 26-35 years Female*

**State**

```python
df.columns
```

```python
#for top 10 states with highest no. of order
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'State', y = 'Orders')
```

```python
#for top 10 states with highest spent on orders
sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'State', y = 'Amount')
```

*Conclusion:: Up, Maharashtra, Karnataka, Delhi, Mp, Ap have the high amount and price ordered*

**Marital Status**

```python
ax = sns.countplot(data = df, x = 'Marital_Status')
sns.set(rc = {'figure.figsize':(7,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set(rc = {'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status', y = 'Amount', hue = 'Gender')
```

*Conclusion:: Married women have highest Purchasing Power*

**Occupation**

```python
sns.set(rc = {'figure.figsize':(22,5)})
ax = sns.countplot(data = df, x = 'Occupation')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set(rc = {'figure.figsize':(22,5)})
sns.barplot(data = sales_state, x = 'Occupation', y = 'Amount')
```

*Conclusion:: Most byers works in IT, Aviation, Healthcare sector*

**Product Category**

```python
sns.set(rc = {'figure.figsize':(27,5)})
ax = sns.countplot(data = df, x = 'Product_Category')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(27,5)})
sns.barplot(data = sales_state, x = 'Product_Category', y = 'Amount')
```

*Conclusion:: Food>Clothing>Electronics are the most sold items*

```python
sales_state = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(27,5)})
sns.barplot(data = sales_state, x = 'Product_ID', y = 'Orders')
#or
fig1, ax1 = plt.subplots(figsize = (12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending = False).plot(kind = 'bar')
```

**Insight::
Married Women of age group 26-35yrs from UP, Maharashtra, Karnataka working in IT, Healthcare, Aviation are more likely buy products from Food, Clothing, Electronics**
